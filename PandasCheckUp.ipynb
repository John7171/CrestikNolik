#Сводники

Анализируем target в нашем датасете. Мы смотрим процент тех кто хочет сменить работу. 1 - да(хочет сменить), 0 - нет.
СТРОИМ СВОДНИКИ ДЛЯ КАТЕГОРИАЛЬНЫХ ПЕРЕМЕННЫХ

```python
df['target'].mean()#мы видим что в столбце target 25% хотят сменить работу. Это тоже сводник но общий
```

с mean пошел анализ, он схлопывает наши 20000 строк DF до 25 %

Смысл в том , что мы можем узнать как переменные влияют на желание сменить работу, как город влияет, зарплата,уровень развитости, пол, опыт. Чтобы это понять мы должны посмотреть на статистику, на цифры. Какой, например, процент DS среди мужчин хочет сменить работу и среди женщин и сравнить. Для этого делаются сводные таблицы

```python
df.groupby('gender')['target'].mean()#c помощью groupby мы делаем это.Передали сюда столбец gender и target и видим процент тех кто
# хочет сменить работу.Mean делает агрегацию, схлопывает по категориям уже, более узко по полу
```

сводная таблица должна содержать count, т.е. количество, в нашем случае число тех людей кто хочет сменить работу. Добавим это в верхнюю строчку кода

```python
t = df.groupby('gender')['target'].agg(['count','mean'])
```

```python
t
```

мы видим что количество 14650 проанализированных, куда-то пропало почти 6000 (тех кто не указал пол). добавляем dropna = False

```python
t['count'].sum()
```

```python
t = df.groupby('gender',dropna = False)['target'].agg(['count','mean'])
```

```python
t#мы видим что среди тех кто не указал пол проыент желающих сменить работу выше
```

Делаем сводник по другой категории, по образованию. Среди graduate наибольший процент желающих сменить работу

```python
df.groupby('education_level',dropna = False)['target'].agg(['count','mean'])
```

добавив mean и plot построили график

```python
df.groupby('education_level',dropna = False)['target'].mean().plot()
```
